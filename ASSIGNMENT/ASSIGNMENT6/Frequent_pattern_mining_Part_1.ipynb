```python
# Importing the libraries
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
# Loading the data
apriori_Gdata = pd.read_csv("Bakery.csv")
apriori_Gdata.product_name = ["".join(i.split(",")) for i in apriori_Gdata.Items]
apriori_Gorder = apriori_Gdata.TransactionNo.unique()
```

```python
 #Transforming the data
apriori_trans = [apriori_Gdata[apriori_Gdata.TransactionNo == i]["Items"].tolist() for i in apriori_Gorder]
apriori_encoder = TransactionEncoder().fit(apriori_trans)
apriori_trans = apriori_encoder.transform(apriori_trans)
apriori_trans = pd.DataFrame(apriori_trans,columns=apriori_encoder.columns_)
```

```python
# Applying apriori algorithm
freq_apriori = apriori(apriori_trans,min_support=0.0095,use_colnames=True)

# Applying association rules
apriori_association_data = association_rules(freq_apriori,metric="lift")
for item in apriori_association_data.iterrows():
  print(f"Rule {item[0] + 1}: {tuple(item[1].antecedents)} -> {tuple(item[1].consequents)}")
```
